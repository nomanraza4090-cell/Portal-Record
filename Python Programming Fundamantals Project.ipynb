# Student Management System

```python
"""
PROGRAM 6: Student Management System (Complete Project)
"""
class StudentManagementSystem:
    """Complete Student Management System"""
    
    def __init__(self):
        self.students = []
        self.next_id = 1
    
    def add_student(self, name, age, marks):
        """Add a new student"""
        student = {
            'id': self.next_id,
            'name': name,
            'age': age,
            'marks': marks,
            'percentage': (marks / 100) * 100,
            'grade': self.calculate_grade(marks)
        }
        self.students.append(student)
        self.next_id += 1
        print(f"Student {name} added successfully!")
        return student
    
    def calculate_grade(self, marks):
        """Calculate grade based on marks"""
        if marks >= 90:
            return 'A+'
        elif marks >= 80:
            return 'A'
        elif marks >= 70:
            return 'B'
        elif marks >= 60:
            return 'C'
        elif marks >= 50:
            return 'D'
        else:
            return 'F'
    
    def display_all_students(self):
        """Display all students"""
        if not self.students:
            print("No students found!")
            return
        
        print("\n" + "="*70)
        print("STUDENT LIST")
        print("="*70)
        print(f"{'ID':<5} {'Name':<20} {'Age':<5} {'Marks':<8} {'Percentage':<12} {'Grade':<6}")
        print("-"*70)
        
        for student in self.students:
            print(f"{student['id']:<5} {student['name']:<20} {student['age']:<5} "
                  f"{student['marks']:<8} {student['percentage']:<12.1f} {student['grade']:<6}")
        print("="*70)
    
    def search_student(self, name):
        """Search student by name"""
        results = [s for s in self.students if name.lower() in s['name'].lower()]
        
        if not results:
            print(f"No student found with name: {name}")
            return
        
        print(f"\nSearch Results for '{name}':")
        for student in results:
            print(f"ID: {student['id']}, Name: {student['name']}, Marks: {student['marks']}, Grade: {student['grade']}")
    
    def update_marks(self, student_id, new_marks):
        """Update student marks"""
        for student in self.students:
            if student['id'] == student_id:
                student['marks'] = new_marks
                student['percentage'] = (new_marks / 100) * 100
                student['grade'] = self.calculate_grade(new_marks)
                print(f"Updated marks for {student['name']} to {new_marks}")
                return
        print(f"Student with ID {student_id} not found!")
    
    def delete_student(self, student_id):
        """Delete a student"""
        for i, student in enumerate(self.students):
            if student['id'] == student_id:
                removed = self.students.pop(i)
                print(f"Removed student: {removed['name']}")
                return
        print(f"Student with ID {student_id} not found!")
    
    def get_statistics(self):
        """Calculate class statistics"""
        if not self.students:
            print("No data available!")
            return
        
        marks_list = [s['marks'] for s in self.students]
        
        print("\n" + "="*60)
        print("CLASS STATISTICS")
        print("="*60)
        print(f"Total Students: {len(self.students)}")
        print(f"Average Marks: {sum(marks_list)/len(marks_list):.2f}")
        print(f"Highest Marks: {max(marks_list)}")
        print(f"Lowest Marks: {min(marks_list)}")
        
        # Grade distribution
        grade_dist = {}
        for student in self.students:
            grade = student['grade']
            grade_dist[grade] = grade_dist.get(grade, 0) + 1
        
        print("\nGrade Distribution:")
        for grade, count in sorted(grade_dist.items()):
            print(f"  {grade}: {count} students")

def run_student_management_system():
    """Main program to run the Student Management System"""
    sms = StudentManagementSystem()
    
    while True:
        print("\n" + "="*60)
        print("STUDENT MANAGEMENT SYSTEM")
        print("="*60)
        print("1. Add Student")
        print("2. Display All Students")
        print("3. Search Student")
        print("4. Update Marks")
        print("5. Delete Student")
        print("6. View Statistics")
        print("7. Exit")
        print("="*60)
        
        choice = input("Enter your choice (1-7): ")
        
        if choice == '1':
            print("\n--- Add New Student ---")
            name = input("Enter student name: ")
            age = int(input("Enter age: "))
            marks = float(input("Enter marks (0-100): "))
            sms.add_student(name, age, marks)
        
        elif choice == '2':
            sms.display_all_students()
        
        elif choice == '3':
            print("\n--- Search Student ---")
            name = input("Enter student name to search: ")
            sms.search_student(name)
        
        elif choice == '4':
            print("\n--- Update Marks ---")
            student_id = int(input("Enter student ID: "))
            new_marks = float(input("Enter new marks (0-100): "))
            sms.update_marks(student_id, new_marks)
        
        elif choice == '5':
            print("\n--- Delete Student ---")
            student_id = int(input("Enter student ID to delete: "))
            sms.delete_student(student_id)
        
        elif choice == '6':
            sms.get_statistics()
        
        elif choice == '7':
            print("\nThank you for using Student Management System!")
            break
        
        else:
            print("Invalid choice! Please try again.")

#  Uncomment to run the system
    run_student_management_system()
```
