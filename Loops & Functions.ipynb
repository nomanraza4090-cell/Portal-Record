# OPERATORS (Arithmetic, Comparison, Logical)

```python
# PROGRAM 2: Operators Demonstration

def operators():
    
    # Arithmetic Operators
    a, b = 15, 4
    print("Addition:", a+b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    print("Division:", a / b)
    print("Floor Division:", a // b)
    print("Modulus:", a % b)
    print("Exponentiation:", a ** b)
    
    print("============================================")
    # Comparison Operators
    print("Equal to:", a == b)
    print(f"Not equal to:", a != b)
    print(f"Greater than:", a > b)
    print(f"Less than:", a < b)
    print(f"Greater than or equal:", a >= b)
    print(f"Less than or equal:", a <= b)
    
    print("============================================")
    # Logical Operators
    p, q = True, False
    print("AND (p and q):", p and q)
    print("OR (p or q):", p or q)
    print("NOT (not p):", not p)
    
    print("============================================")
    # Assignment Operators
    num = 10
    print("Initial value: {num}")
    num += 5; print("After +=5:", num)
    num -= 3; print("After -=3:", num)
    num *= 2; print("After *=2:", num)
    num /= 4; print("After /=4:", num)

operators()
```

# CONDITIONAL STATEMENTS (if-elif-else)

```python
# PROGRAM 3: Conditional Statements

def conditional_statements():
    
    # Grade Calculator
    marks = float(input("Enter your marks: "))
    
    if marks >= 90:
        grade = 'A+'
        remarks = "Excellent!"
    elif marks >= 80:
        grade = 'A'
        remarks = "Very Good!"
    elif marks >= 70:
        grade = 'B'
        remarks = "Good!"
    elif marks >= 60:
        grade = 'C'
        remarks = "Satisfactory"
    elif marks >= 50:
        grade = 'D'
        remarks = "Need Improvement"
    else:
        grade = 'F'
        remarks = "Fail. Please work harder!"
    
    print("Marks: {marks}")
    print("Grade: {grade}")
    print("Remarks: {remarks}")
    
    # Even/Odd Check
    number = int(input("Enter a number: "))
    
    if number % 2 == 0:
        print(f"{number} is EVEN")
    else:
        print(f"{number} is ODD")
    
    # Voting Eligibility
    print("\n3. VOTING ELIGIBILITY:")
    age = int(input("Enter your age: "))
    
    if age >= 18:
        print("You are eligible to vote!")
        if age >= 60:
            print("You are a senior citizen voter.")
    else:
        print(f"You need to wait {18 - age} more years to vote.")

conditional_statements()
```

#  LOOPS (for and while)

```python
# PROGRAM 4: Loops (for and while)

def loops():
    
    # For Loop 
    
    # Print 1 to 10
    print("numbers from 1 to 10:")
    for i in range(1, 11):
        print(i, end=" ")
    print()
    
    print("============================================")
    # Sum of first N numbers
    n = 10
    sum_numbers = 0
    for i in range(1, n+1):
        sum_numbers += i
    print(f"Sum of first {n} numbers: {sum_numbers}")
    

    # Multiplication Table
    print("\nMultiplication Table of 5:")
    num = 5
    for i in range(1, 11):
        print(f"{num} × {i} = {num * i}")
    
    print("============================================")
    # Iterate through list
    print("\nIterating through list:")
    fruits = ["Apple", "Banana", "Orange", "Mango", "Grapes"]
    for fruit in fruits:
        print(f"I like {fruit}")
    
    # While Loop 
    print("============================================")
    # Countdown
    print("Countdown:")
    count = 5
    while count > 0:
        print(count)
        count -= 1 
    print("Blast Off!")
    
    print("============================================")
    # Factorial
    print("\nFactorial Calculator:")
    number = int(input("Enter a number: "))
    factorial = 1
    i = 1
    while i <= number:
        factorial *= i
        i += 1
    print(f"Factorial of {number} is {factorial}")
    
    print("============================================")
    # Nested Loops
    print("\n3. NESTED LOOPS:")
    print("Pattern Printing:")
    rows = 5
    for i in range(1, rows+1):
        for j in range(1, i+1):
            print("*", end=" ")
        print()

loops()
```

# FUNCTIONS

```python
# PROGRAM 5: Functions

# Basic Function
def greet(name):
    return f"Hello, {name}! Welcome to Python programming."

# Function with multiple parameters
def calculate_area(length, width):
    return length * width

# Function with default parameter
def power(base, exponent=2):
    return base ** exponent

# Function with variable arguments
def sum_all(*numbers):
    return sum(numbers)

# Recursive function
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

# Lambda function
square = lambda x: x ** 2

# Function returning multiple values
def get_student_info():
    name = "Alice"
    age = 22
    grade = "A"
    return name, age, grade

def demonstrate_functions():
    # Call basic function
    print(greet("Nomi"))
    
    # Call function with parameters
    area = calculate_area(5, 4)
    print(f"Area of rectangle 5*4: {area}")
    
    # Call function with default parameter
    print(f"2^2 (default): {power(2)}")
    print(f"2^3 (custom): {power(2, 3)}")
    print("============================================")
    # Variable arguments
    result = sum_all(1, 2, 3, 4, 5)
    print(f"Sum of 1,2,3,4,5: {result}")
    
    # Recursive function
    num = 5
    print(f"Factorial of {num}: {factorial(num)}")
    
    print("============================================")
    # Lambda function
    numbers = [1, 2, 3, 4, 5]
    squares = list(map(lambda x: x**2, numbers))
    print(f"Original: {numbers}")
    print(f"Squares: {squares}")
    
    print("============================================")
    # Multiple return values
    name, age, grade = get_student_info()
    print(f"Name: {name}, Age: {age}, Grade: {grade}")
    
    print("============================================")
    # Function documentation
    print(f"greet function doc: {greet.__doc__}")
    print(f"calculate_area doc: {calculate_area.__doc__}")

demonstrate_functions()
```
