```python
# Part 1 & 2: FizzBuzz from 1 to 20
print("FizzBuzz from 1 to 20:")
for num in range(1, 21):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)

print() 

user_number = int(input("Enter a number for countdown: "))


print(f"\nCountdown from {user_number}:")
count = user_number

while count >= 1:
  
    if user_number > 10:
        print(f"Number {user_number} is greater than 10. Countdown stopped!")
        break
    
    print(count)
    count -= 1
    if count == 0 and user_number <= 10:
        print("Blast Off!")
```
