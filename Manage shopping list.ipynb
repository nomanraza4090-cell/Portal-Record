# 1. Create a list named shopping_list containing at least 5 initial items 

```python
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]
```

# 2. Print the entire list, and also print the third item in the list.

```python

print("Initial shopping list:", shopping_list)
print("Third item in the list:", shopping_list[2])
print()
```

# 3. Add two new items to the list using an appropriate list method.

```python
shopping_list.append("rice")
shopping_list.append("yogurt")
print("After adding two items:", shopping_list)
```

# 4. Remove one item from the list.

```python
shopping_list.remove("eggs")
print("After removing one item:", shopping_list)
print()
```

# 5. Create a dictionary named item_prices.

```python
item_prices = {
    "milk": 200,
    "bread": 170,
    "apples": 200,
    "chicken": 400,
    "rice": 120
}
print("Price dictionary:", item_prices)
```

# 6. Print the price of one specific item from the dictionary.

```python
print("Price of bread: Rs: ", item_prices["bread"])
print()
```

# 7. Print a formatted summary of your final shopping list and the price dictionary. 

```python
print("=" * 50)
print("FINAL SHOPPING SUMMARY")
print("=" * 50)
print("Shopping List Items:")
for item in shopping_list:
    print(f"  • {item}")

print("\nItem Prices:")
for item, price in item_prices.items():
    print(f"  • {item}: Rs{price:.2f}")

print("=" * 50)
```
