1. Variables (Name, Age, City)

```python
name = "Momin"
age = 21
city = "Gujranwala"

print(name, age, city)
```

2. Data Types (int, float, string, boolean)

```python
num = 10
decimal = 10.5
text = "Hello"
is_true = True

print(num, decimal, text, is_true)
```

3. Check Data Types using type()

```python
print(type(num))
print(type(decimal))
print(type(text))
print(type(is_true))
```

 4. Arithmetic Operations

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

5. Comparison Operators

```python
x = 10
y = 20

print(x > y)   # Greater than
print(x < y)   # Less than
print(x == y)  # Equal to
```

6. Type Conversion

```python
# Integer to float
num_int = 10
num_float = float(num_int)
print(num_float)

# Float to integer
num_float2 = 10.75
num_int2 = int(num_float2)
print(num_int2)
```

7. Student Information Program

```python
name = "Ali"
roll_no = 101
marks = 85

print("Name:", name)
print("Roll Number:", roll_no)
print("Marks:", marks)
```

8. Assignment Operators

```python
a = 10

a += 5
print("After += :", a)

a -= 3
print("After -= :", a)

a *= 2
print("After *= :", a)
```

9. Boolean Operations

```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

10. Area of Rectangle

```python
length = 5
width = 3

area = length * width
print("Area of rectangle:", area)
```
