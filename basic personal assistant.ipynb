# 1. basic personal assistant

```python
name = input("What is your name? ")
```

# 2. Ask the user for their birth year

```python
birth_year = int(input("What year were you born? "))
```

# 3. Ask the user for their current age

```python
current_age = int(input("What is your current age? "))
```

# 4. Calculate and print the year they will turn 100 years old.

```python
current_year = 2026 
year_turn_100 = current_year + (100 - current_age)
print(f"\nHello {name}! You will turn 100 years old in the year {year_turn_100}.")
```

# 5. Ask the user for a lucky number

```python
lucky_number = float(input("Enter your lucky number: "))
```

# 6. Print a summary message that includes all the information you've collected

```python
print("\n" + "="*50)
print("USER INFORMATION SUMMARY")
print("="*50)


print(f"Name '{name}' is of type: {type(name)}")
print(f"Birth Year {birth_year} is of type: {type(birth_year)}")
print(f"Current Age {current_age} is of type: {type(current_age)}")
print(f"Lucky Number {lucky_number} is of type: {type(lucky_number)}")

print("\n" + "="*50)
print("TYPECASTING EXAMPLES")
print("="*50)

age_as_float = float(current_age)
print(f"Age as float: {age_as_float} (Type: {type(age_as_float)})")

lucky_as_integer = int(lucky_number)
print(f"Lucky number as integer: {lucky_as_integer} (Type: {type(lucky_as_integer)})")


birth_year_as_string = str(birth_year)
print(f"Birth year as string: '{birth_year_as_string}' (Type: {type(birth_year_as_string)})")
```

# 7.  Add comments to your code explaining each step.

```python
print("\n" + "="*50)
print("CALCULATION DETAILS")
print("="*50)

print(f"Calculation: Current Year ({current_year}) + (100 - Current Age ({current_age})) = {year_turn_100}")
print(f"Therefore, {name} will be 100 years old in {year_turn_100}")

calculated_age = current_year - birth_year
print(f"\nAge verification: Based on birth year, you should be approximately {calculated_age} years old.")
if calculated_age == current_age:
    print("✓ Your birth year and age match perfectly!")
else:
    print("Note: There might be a slight difference if you haven't had your birthday yet this year.")

print("="*50)
```
