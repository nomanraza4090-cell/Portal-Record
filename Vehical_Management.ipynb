```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    def __init__(self,licence_plate,max_speed):
        self.__engine_status=False
        self.__licence_plate=licence_plate
        self.__max_speed=max_speed
    @abstractmethod
    def start_engine(self):
        pass
    @abstractmethod
    def stop_engine(self):
        pass
    @abstractmethod
    def get_details(self):
        pass
    def set_engine_status(self,status):
        self.__engine_status=status
    def get_engine_status(self):
        return self.__engine_status
    def set_licence_plate(self,licence_plate):
        self.__licence_plate=licence_plate
    def get_licence_plate(self):
        return self.__licence_plate
    def set_max_speed(self,speed):
        if speed > 0:
            self.__max_speed=speed
        else:
            print("Enter possitive value!")
    def get_max_speed(self):
        return self.__max_speed
    
```

```python
class Car(Vehicle):
    def start_engine(self):
        self.set_engine_status(True)
        print(f"Car :{self.get_licence_plate()}, Engine Started.")
    def stop_engine(self):
        self.set_engine_status(False)
        print(f"Car :{self.get_licence_plate()}, Engine stopped.")
    def get_details(self):
        print(f"Vehicle's detail -> Car-Plate :{self.get_licence_plate()}  ,  Max-Speed :{self.get_max_speed()}km/h")
```

```python
    
class Truck(Vehicle):
    def start_engine(self):
        self.set_engine_status(True)
        print(f"Truck :{self.get_licence_plate()}, Engine started.")
    def stop_engine(self):
        self.set_engine_status(False)
        print(f"Truck :{self.get_licence_plate()}, Engine stopped.")
    def get_details(self):
        print(f"Vehicle's detail -> Truck-Plate :{self.get_licence_plate()}  ,  Max-Speed :{self.get_max_speed()}km/h")
```

```python

def test_vehicle():
    vehicles =[
        Car("GAN 1235",230),
        Truck("APL 7876",150),
        Car("GAN 572",200),
        Truck("GAL 7832",180)    
    ]

    for vehicle in vehicles:
        vehicle.start_engine()
        vehicle.get_details()
        vehicle.stop_engine()
        print ("-" *60)

if __name__=="__main__":
    test_vehicle()
```
