```python
# Base class Person
class Person:
    def __init__(self, name, age):
       
        self.name = name
        self.age = age
    
    def display_info(self):
        
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


# Derived class Student inheriting from Person
class Student(Person):
    def __init__(self, name, age, student_id, major):
        
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major
    
    def display_info(self):
        
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")


# Derived class GraduateStudent inheriting from Student
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
       
        # Call parent class constructor
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic
    
    def display_info(self):
      
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")
        print(f"Thesis Topic: {self.thesis_topic}")


# Create objects and test the classes
def main():
    print("=" * 50)
    print("OOP CLASS SYSTEM DEMONSTRATION")
    print("=" * 50)
    
    # Create Person object
    print("\n--- Person ---")
    person1 = Person("John Doe", 45)
    person1.display_info()
    
    # Create Student object
    print("\n--- Student ---")
    student1 = Student("Jane Smith", 20, "S12345", "Computer Science")
    student1.display_info()
    
    # Create GraduateStudent object
    print("\n--- Graduate Student ---")
    grad_student1 = GraduateStudent(
        "Bob Johnson", 
        27, 
        "G67890", 
        "Physics", 
        "Quantum Computing"
    )
    grad_student1.display_info()
    
    print("\n" + "=" * 50)
    print("Demonstration of Polymorphism")
    print("=" * 50)
    
    # Demonstrate polymorphism
    persons = [
        Person("Alice Brown", 35),
        Student("Charlie Wilson", 22, "S98765", "Mathematics"),
        GraduateStudent("Diana Prince", 28, "G54321", "Biology", "Genetics Research")
    ]
    
    print("\n--- Polymorphism Example ---")
    for person in persons:
        print("\n--- " + person.__class__.__name__ + " ---")
        person.display_info()
        print("-" * 30)


# Run the program
if __name__ == "__main__":
    main()
```
